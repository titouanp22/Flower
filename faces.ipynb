```python
from pathlib import Path
import sys
import urllib.request
import random
import json

import numpy as np
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt

from torchvision import transforms
from torchvision.datasets import ImageFolder


from sklearn.datasets import fetch_olivetti_faces
from PIL import Image
```

```python
# Locate Flower root (fixed: local cloned repo in this project)
cwd = Path.cwd().resolve()
flow_root_candidate = cwd / "mehrsapo_Flower"
if not flow_root_candidate.exists():
    flow_root_candidate = cwd.parent / "mehrsapo_Flower"

if not ((flow_root_candidate / "pnpflow").exists() and (flow_root_candidate / "config").exists()):
    raise RuntimeError(f"Impossible de trouver le repo Flower attendu: {flow_root_candidate}")

FLOWER_ROOT = flow_root_candidate

if str(FLOWER_ROOT) not in sys.path:
    sys.path.insert(0, str(FLOWER_ROOT))

print("FLOWER_ROOT:", FLOWER_ROOT)
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("DEVICE:", DEVICE)
```

```python
from pnpflow.models import UNet

from scripts.faces_pipeline import LocalSuperresolution, SingleImageDataset, psnr_torch, set_seed
from scripts.flower_steps import run_flower_inverse_problem_steps
```

```python
# Load pretrained Flower weights (CelebA OT)
weights_path = FLOWER_ROOT / "model" / "celeba" / "ot" / "model_final.pt"
weights_path.parent.mkdir(parents=True, exist_ok=True)
if not weights_path.exists():
    url = "https://huggingface.co/mehrsapo/flower-weights/resolve/main/model/celeba/ot/model_final.pt"
    print("Downloading:", url)
    urllib.request.urlretrieve(url, weights_path)

model = UNet(
    input_channels=3,
    input_height=128,
    ch=32,
    ch_mult=(1, 2, 4, 8),
    num_res_blocks=6,
    attn_resolutions=(16, 8),
    resamp_with_conv=True,
).to(DEVICE)

state = torch.load(weights_path, map_location=DEVICE)
model.load_state_dict(state)
model.eval()
print("Loaded model from:", weights_path)
```

```python
# Load a single custom face image (no dataset needed)
WORK_DIR = Path.cwd().resolve() / "results" / "faces"
DATA_DIR = WORK_DIR / "data"
OUT_DIR = WORK_DIR
DATA_DIR.mkdir(parents=True, exist_ok=True)
OUT_DIR.mkdir(parents=True, exist_ok=True)

transform_128 = transforms.Compose([
    transforms.Resize(135),
    transforms.Lambda(lambda im: transforms.functional.affine(
        im, angle=0, translate=(0, 12), scale=1.0, shear=0
    )),
    transforms.CenterCrop(128),
    transforms.ToTensor(),
])

custom_image_path = Path(r"C:/Users/titou/OneDrive/Bureau/mva/gen_images/flower/Flower/experiments/super_resolution_imagenette/data/titouan.jpeg")
if not custom_image_path.exists():
    raise FileNotFoundError(f"Image not found: {custom_image_path}")

img = Image.open(custom_image_path).convert("RGB")
x_gt = transform_128(img).unsqueeze(0).to(DEVICE)

val_ds = SingleImageDataset(img)
print('Using custom image:', custom_image_path)
print('shape:', tuple(x_gt.shape))
print('Results will be saved in:', OUT_DIR)
```

```python
# SR setup
SF = 2
SIGMA_NOISE = 0
N_STEPS = 500
OBSERVE_STEPS = [10, 50, 100]
SEED = 0

set_seed(SEED)
```

```python
degradation = LocalSuperresolution(sf=SF)
H = degradation.H
Ht = degradation.H_adj
y = H(x_gt) + SIGMA_NOISE * torch.randn_like(H(x_gt))
```

```python
# Show HR (x_gt), LR (y), and LR mapped back to HR grid (Ht(y))
x_hr = x_gt[0].detach().cpu().permute(1, 2, 0).numpy()
x_lr = y[0].detach().cpu().permute(1, 2, 0).numpy()
x_lr_up = Ht(y)[0].detach().cpu().permute(1, 2, 0).numpy()

fig, ax = plt.subplots(1, 3, figsize=(16, 4))

ax[0].imshow(np.clip(x_hr, 0, 1))
ax[0].set_title("HR (Ground Truth)")
ax[0].axis("off")

ax[1].imshow(np.clip(x_lr, 0, 1), interpolation='nearest')
ax[1].set_title("LR (pixelized)")
ax[1].axis("off")

ax[2].imshow(np.clip(x_lr_up, 0, 1))
ax[2].set_title("LR mapped to HR grid: Ht(y)")
ax[2].axis("off")

plt.tight_layout()
plt.show()
```

```python
# Flower loop with explicit steps
x_final, details = run_flower_inverse_problem_steps(
    model=model,
    y=y,
    H=H,
    Ht=Ht,
    sigma_noise=SIGMA_NOISE,
    N=N_STEPS,
    observe_steps=OBSERVE_STEPS,
)
print('done. observed:', sorted(details.keys()))
```

```python
# Visualize GT / input / final + step-by-step
x_gt_np = x_gt[0].detach().cpu().permute(1,2,0).numpy()
y_up = Ht(y)
y_up_np = y_up[0].detach().cpu().permute(1,2,0).numpy()
x_final_np = x_final[0].permute(1,2,0).numpy()

fig, ax = plt.subplots(1, 3, figsize=(12, 4))
ax[0].imshow(np.clip(x_gt_np, 0, 1)); ax[0].set_title('Ground truth')
ax[1].imshow(np.clip(y_up_np, 0, 1)); ax[1].set_title('Input (adjoint)')
ax[2].imshow(np.clip(x_final_np, 0, 1)); ax[2].set_title('Flower final')
for a in ax:
    a.axis('off')
plt.tight_layout()
plt.show()

rows = len(details)
fig2, axes = plt.subplots(rows, 4, figsize=(14, 3.0 * rows), squeeze=False)
cols = ['x_t', 'Step 1: x_hat', 'Step 2: x_star', 'Step 3: x_next']
for j, c in enumerate(cols):
    axes[0][j].set_title(c)

for i, step in enumerate(sorted(details.keys())):
    d = details[step]
    imgs = [d['x_t'], d['x_hat'], d['x_star'], d['x_next']]
    for j, ten in enumerate(imgs):
        arr = ten[0].permute(1,2,0).numpy()
        axes[i][j].imshow(np.clip(arr, 0, 1))
        axes[i][j].axis('off')
        if j == 0:
            axes[i][j].set_ylabel(f"step={step} t={d['t']:.2f}")

plt.tight_layout()
out_steps = OUT_DIR / 'flower_faces_steps.png'
plt.savefig(out_steps, dpi=160)
plt.show()
print('saved:', out_steps)

# PSNR over time (using x_next at each step)
steps_sorted = sorted(details.keys())
psnr_vals = []
t_vals = []
for step in steps_sorted:
    d = details[step]
    x_step = torch.clamp(d['x_next'].to(x_gt.device), 0, 1)
    psnr_vals.append(float(psnr_torch(x_step, x_gt)))
    t_vals.append(float(d['t']))

fig_psnr, ax_psnr = plt.subplots(1, 1, figsize=(6, 3.5))
ax_psnr.plot(t_vals, psnr_vals, marker='o', lw=1.5)
ax_psnr.set_xlabel('t')
ax_psnr.set_ylabel('PSNR (dB)')
ax_psnr.set_title('PSNR over time')
ax_psnr.grid(True, alpha=0.3)
plt.tight_layout()
out_psnr = OUT_DIR / 'flower_faces_psnr_over_time.png'
plt.savefig(out_psnr, dpi=160)
plt.show()
print('saved:', out_psnr)
```

```python
# PSNR + baseline + small gallery

x_input_adjoint = Ht(y)
x_input_bicubic = F.interpolate(y, size=x_gt.shape[-2:], mode='bicubic', align_corners=False)
x_flower = x_final.to(x_gt.device)

psnr_adj = psnr_torch(torch.clamp(x_input_adjoint, 0, 1), x_gt)
psnr_bic = psnr_torch(torch.clamp(x_input_bicubic, 0, 1), x_gt)
psnr_flw = psnr_torch(torch.clamp(x_flower, 0, 1), x_gt)
print(f'PSNR adjoint: {psnr_adj:.2f} dB')
print(f'PSNR bicubic: {psnr_bic:.2f} dB')
print(f'PSNR Flower : {psnr_flw:.2f} dB | gain vs bicubic: {psnr_flw - psnr_bic:+.2f} dB')

# mini-gallery on a few face indices (with low-resolution degraded input)
indices = list(range(min(4, len(val_ds))))
rows = []
for id_i in indices:
    img_i, _ = val_ds[id_i]
    gt_i = transform_128(img_i).unsqueeze(0).to(DEVICE)
    y_i = H(gt_i) + SIGMA_NOISE * torch.randn_like(H(gt_i))
    lr_vis_i = F.interpolate(y_i, size=gt_i.shape[-2:], mode='nearest')
    bic_i = F.interpolate(y_i, size=gt_i.shape[-2:], mode='bicubic', align_corners=False)
    out_i, _ = run_flower_inverse_problem_steps(model, y_i, H, Ht, SIGMA_NOISE, min(N_STEPS, 60), observe_steps=[])
    out_i = out_i.to(DEVICE)
    rows.append((
        gt_i[0].cpu(),
        lr_vis_i[0].cpu(),
        bic_i[0].cpu(),
        out_i[0].cpu(),
        psnr_torch(bic_i, gt_i),
        psnr_torch(out_i, gt_i),
    ))

fig, axes = plt.subplots(len(rows), 4, figsize=(12, 3 * len(rows)), squeeze=False)
for i, (gt_i, lr_i, inp_i, out_i, p_b, p_f) in enumerate(rows):
    axes[i, 0].imshow(np.clip(gt_i.permute(1,2,0).numpy(), 0, 1)); axes[i, 0].set_title('GT'); axes[i, 0].axis('off')
    axes[i, 1].imshow(np.clip(lr_i.permute(1,2,0).numpy(), 0, 1)); axes[i, 1].set_title('Low-res'); axes[i, 1].axis('off')
    axes[i, 2].imshow(np.clip(inp_i.permute(1,2,0).numpy(), 0, 1)); axes[i, 2].set_title(f'Bicubic\nPSNR={p_b:.2f}'); axes[i, 2].axis('off')
    axes[i, 3].imshow(np.clip(out_i.permute(1,2,0).numpy(), 0, 1)); axes[i, 3].set_title(f'Flower\nPSNR={p_f:.2f}'); axes[i, 3].axis('off')

plt.tight_layout()
out_cmp = OUT_DIR / 'flower_faces_vs_bicubic_examples.png'
plt.savefig(out_cmp, dpi=160)
plt.show()
print('saved:', out_cmp)
```

```python
# Same image, dense timeline + reference columns (GT / LR / Step1 / Step2 / Step3)

# Choose image index
N_SHOW = 20

img, cls = val_ds[0]


obs_dense = sorted(set(np.linspace(1, N_STEPS, N_SHOW, dtype=int).tolist()))

set_seed(SEED)

x_final_dense, details_dense = run_flower_inverse_problem_steps(
    model=model,
    y=y,
    H=H,
    Ht=Ht,
    sigma_noise=SIGMA_NOISE,
    N=N_STEPS,
    observe_steps=obs_dense,
)

# References
gt_ref = x_gt[0].detach().cpu().permute(1, 2, 0).numpy()               # HR
lr_ref = y[0].detach().cpu().permute(1, 2, 0).numpy()                  # true LR (small)
lr_up_ref = Ht(y)[0].detach().cpu().permute(1, 2, 0).numpy()           # LR mapped to HR grid

rows = len(details_dense)
fig, axes = plt.subplots(rows, 5, figsize=(16, 2.8 * rows), squeeze=False)
col_titles = ["HR (GT)", "LR upsampled", "Step 1: x_hat", "Step 2: x_star", "Step 3: x_next"]
for j, title in enumerate(col_titles):
    axes[0][j].set_title(title)

for i, step in enumerate(sorted(details_dense.keys())):
    d = details_dense[step]

    # same references repeated on each row
    axes[i][0].imshow(np.clip(gt_ref, 0, 1))
    axes[i][1].imshow(np.clip(lr_up_ref, 0, 1))

    imgs = [d["x_hat"], d["x_star"], d["x_next"]]
    for j, ten in enumerate(imgs, start=2):
        arr = ten[0].permute(1, 2, 0).numpy()
        axes[i][j].imshow(np.clip(arr, 0, 1))

    for j in range(5):
        axes[i][j].axis("off")

    axes[i][0].set_ylabel(f"step={step}  t={d['t']:.2f}")

plt.tight_layout()
out_dense = OUT_DIR / "flower_pretrained_superres_steps_dense_with_refs.png"
plt.savefig(out_dense, dpi=160)
plt.show()
print("saved:", out_dense)
```
